```python
import pandas as pd
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Load your dataset (replace 'your_dataset.csv' with the actual file path)
data = pd.read_csv('Final Transactions.csv')
```

```python
# Data Cleansing
data = data.drop_duplicates()  # Remove duplicate entries
data = data.dropna()  # Remove rows with missing values
# You can add more specific data cleansing steps based on your dataset
```

```python
data
```

```python
# Feature Engineering
data['TX_FREQUENCY'] = data.groupby('CUSTOMER_ID')['TX_DATETIME'].transform('count')
data['TX_FREQUENCY']
# Add more feature engineering steps based on your research goals
```

```python
from scipy.stats import pearsonr
data['TX_TIMESTAMP'] = pd.to_datetime(data['TX_DATETIME'])
data['DAY_OF_WEEK'] = data['TX_TIMESTAMP'].dt.dayofweek
data['HOUR_OF_DAY'] = data['TX_TIMESTAMP'].dt.hour

# Calculate dynamic correlations between temporal attributes
dynamic_correlations = {}
temporal_attributes = ['DAY_OF_WEEK', 'HOUR_OF_DAY']  # Add more temporal attributes as needed
for attr1 in temporal_attributes:
    dynamic_correlations[attr1] = {}
    for attr2 in temporal_attributes:
        corr, _ = pearsonr(data[attr1], data[attr2])
        dynamic_correlations[attr1][attr2] = corr

# Construct a correlation matrix
correlation_matrix = pd.DataFrame(dynamic_correlations)

# Display the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
from sklearn.preprocessing import MinMaxScaler
# Temporal Segmentation
# Implement temporal segmentation based on TX_TIME_SECONDS
window_duration = 3600  # Define the window duration in seconds
data['TX_WINDOW'] = (data['TX_TIME_SECONDS'] // window_duration) * window_duration

# Scaling
scaler = MinMaxScaler()
data[['TX_AMOUNT', 'TX_FREQUENCY']] = scaler.fit_transform(data[['TX_AMOUNT', 'TX_FREQUENCY']])



```

```python
from imblearn.over_sampling import SMOTE

# Split the dataset into features (X) and target (y)
X = data[['TX_AMOUNT', 'DAY_OF_WEEK','HOUR_OF_DAY','TX_TIME_SECONDS','TX_FREQUENCY','TX_WINDOW']]  # Replace with relevant features
y = data['TX_FRAUD']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=4587655)

# Apply SMOTE for oversampling the minority class
smote = SMOTE(random_state=42)
X_train, y_train = smote.fit_resample(X_train, y_train)

# Train an XGBoost classifier
model = XGBClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate performance metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Print the results
print("Precision:", precision)
print("Recall:", recall)
print("F1:", f1)
print("accuracy:", accuracy)
```

```python
# Temporal Correlation Analysis (Example: Time-based visualization)
fraud_predictions = model.predict(X)  # Predict fraud for the entire dataset
data['FRAUD_PREDICTION'] = fraud_predictions

# Visualize temporal correlation
import matplotlib.pyplot as plt

fraud_data = data[data['TX_FRAUD'] == 1]  # Actual fraudulent transactions
plt.figure(figsize=(10, 6))
plt.scatter(fraud_data['TX_TIME_SECONDS'], fraud_data['TX_AMOUNT'], color='red', label='Fraudulent')
plt.scatter(data['TX_TIME_SECONDS'], data['TX_AMOUNT'], color='blue', alpha=0.1, label='All Transactions')
plt.xlabel('Transaction Time (seconds)')
plt.ylabel('Transaction Amount')
plt.title('Temporal Correlation Analysis')
plt.legend()
plt.show()
```

```python
# Visualize the correlation matrix (Temporal Interpretability Score - TIS)
plt.figure(figsize=(10, 6))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar(label='Pearson Correlation')
plt.title('Temporal Interpretability Score (TIS)')
plt.xticks(range(len(temporal_attributes)), temporal_attributes, rotation=45)
plt.yticks(range(len(temporal_attributes)), temporal_attributes)
plt.tight_layout()
plt.show()
```
